```python
rows = soup.find_all("div", class_="sc-beb003d5-2 bkNrIb")
```

```python
data = []
for row in rows:
    name = row.find(class_="sc-4984dd93-0 kKpPOn").text.strip()
    price = row.find(class_="sc-8497df48-2 gtUrBF").text.strip()
    percent= row.find(class_="sc-97d6d2ca-0 bQjSqS").text.strip()
    Market_Cap = row.find(class_="sc-edc9a476-0 fXzXSk").text.strip()
    volume = row.find(class_="sc-aef7b723-0 sc-3f597112-0 eRFZOw").text.strip()
    Circulating_Supply = row.find(class_="sc-4984dd93-0 WfVLk").text.strip()
    
    rich = {
        "name": name,
        "price": price,
        "percent": percent,
        "Market_Cap":Market_Cap,
        "volume":volume,
        "Circulating_Supply":Circulating_Supply
    }
    
    data.append(rich)
```
